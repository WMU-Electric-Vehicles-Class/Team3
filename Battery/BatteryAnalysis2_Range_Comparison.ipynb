## This code is for the battery analysis. The inputs will be different cruising velocities and discharge rates, and the outputs will be maximum vehicle range.
 
> ### Speed (m/s) = v
> ### Nominal voltage (V) = V
> ### Specific energy (Wh/kg) = SE
> ### Battery Energy Capacity (Wh) = E
> ### Mass (kg) = m
> ### Discharged rate (A) = I
> ### Coulometric capacity (Ah) = CC
> ### Fully charged state of charge (%) = SOC1
> ### Minimum state of charge (%) = SOC2
> ### Time to discharge (h) = t
> ### Distance (km) = d


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
SOC1=1
SOC2=.0
I=np.linspace(35,100,50)
v=[15,18, 21, 24]
V=356
SE=147.7
m=457
E=SE*m
CC=E/V
t=((SOC1-SOC2)*CC)/I 
d=[]
```

```python
for velocities in v:
    d.append(velocities*3.6*t)
for numbers in range(0,len(d)):
    plt.plot(I,d[numbers])
plt.xlabel("Discharge rate (Amps)")
plt.ylabel("Range (km)")
plt.title("Discharge Rate (Amps) versus Range of the Vehicle (km)")
plt.grid()
plt.legend(['15 m/s','18 m/s','21 m/s','24 m/s'])
plt.savefig('BatteryPlot12.png',dpi=1000)
plt.show()
```
