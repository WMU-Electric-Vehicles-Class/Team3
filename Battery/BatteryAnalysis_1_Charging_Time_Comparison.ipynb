## This code is for the battery analysis. The inputs will be two different charge rates (for two different chargers) and state of charge and the output will be time required to charge the battery. Variables used:
> ### Nominal Voltage of Battery (V) = V

> ### Nominal Capacity of the Battery (Ah) = C

> ### Time required to charge the battery (h) = t_charge

> ### Charge Rate of the Battery (A) = I 

> ### State of Charge of the Battery (%) = SoC

```python
import numpy
import matplotlib.pyplot as plt


V = 356
C = 189.6
I = [(7200/240), (10500/240)] #Two charge rates as we are examining two different chargers
SOC = numpy.linspace(0,100,25) 
t_charge_level1 = ((1-0.01*SOC)*C)/I[0]
t_charge_level2 = ((1-0.01*SOC)*C)/I[1]
plt.plot(t_charge_level1,SOC)
plt.plot(t_charge_level2,SOC)
plt.grid()
plt.xlabel('Time to Charge (Hours)')
plt.ylabel('State of Charge (%)')
plt.title('Charging time comparison using two different chargers')
plt.legend(['7.2 kW','10.5 kW'])
plt.savefig('BatteryPlot11.png',dpi=1000)
plt.show()
```
