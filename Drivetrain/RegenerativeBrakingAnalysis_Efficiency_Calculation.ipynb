## This code is for regenerative braking analysis. The inputs will be different velocities and the outputs will be regenerative energy efficiency at those velocities.
> ### Speed (km/hr) = v
> ### Nominal Voltage (V) = V
> ### Kinetic Energy (J) = KE
> ### Max Discharge Rate (A) = I
> ### Time(sec) = t
> ### Acceleration (m/s^2) = a
> ### Efficiency (%) = n
> ### Battery Energy (Wh) = Eb

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
I=(10500/240)
velocity=[42,97,145,163] 
V=(356)
m=1657
a=(-5)
n=0.88
y=[]
nr=[]
```

```python
for v in velocity:
    KE=(.5)*m*((v/3.6)**2)
    t=(0-(v/3.6))/a
    Pbmax=V*I
    tpow=(t-(Pbmax/(n*m*(a)**2)))
    Eb=(Pbmax*tpow)+n*m*a*(a/2)*((t)**2-(tpow)**2)
    Nr=float((Eb/KE)*100)
    nr.append(Nr)
```

```python
plt.plot(velocity, nr)
plt.xlabel("Velocity (km/hr)")
plt.ylabel("Efficency (%)")
plt.title("Regenerative Energy Efficency(%) over Velocity (km/hr)")
plt.grid()
plt.savefig('RegenerativeBrakingPlot1.png',dpi=1000)
plt.show()
```
