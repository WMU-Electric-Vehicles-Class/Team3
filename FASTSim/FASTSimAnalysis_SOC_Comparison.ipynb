## This code is used for FASTSim analysis. This code only runs if you have FASTSim installed in your computer. For more information about FASTSim, visit the following link:
https://www.nrel.gov/transportation/fastsim.html#:~:text=The%20Future%20Automotive%20Systems%20Technology,%2C%20cost%2C%20and%20battery%20life.

## We choose a similar electric vehicle - 2016 Chevrolet Spark for this analysis.

```python
from fastsim import vehicle, cycle, simdrive
import matplotlib.pyplot as plt
veh = vehicle.Vehicle(18)
print(veh.Scenario_name)
cyc1 = cycle.Cycle("udds")
cyc2=cycle.Cycle("hwfet")
cyc3=cycle.Cycle("us06")
sim1 = simdrive.SimDriveClassic(cyc1, veh)
sim1.sim_drive()
sim2 = simdrive.SimDriveClassic(cyc2, veh)
sim2.sim_drive()
sim3 = simdrive.SimDriveClassic(cyc3, veh)
sim3.sim_drive()

plt.plot(cyc1.cycSecs,sim1.soc*100,label='UDDS')
plt.plot(cyc2.cycSecs,sim2.soc*100,label='HWFET')
plt.plot(cyc3.cycSecs,sim3.soc*100,label='US06')
plt.xlim(0,800)
plt.ylabel("State of Charge (%)")
plt.xlabel('Time (sec)')
plt.title('SoC Comparison for three different drive-cycles.')
plt.legend()
plt.savefig('DriveCycleComparison.png')
plt.show()
```
