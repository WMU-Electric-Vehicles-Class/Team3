```python
from fastsim import vehicle, cycle, simdrive
import matplotlib.pyplot as plt
veh1 = vehicle.Vehicle(18)
veh2 = vehicle.Vehicle(19)
veh3 = vehicle.Vehicle(20)
cyc1 = cycle.Cycle("udds")
```

```python
sim1 = simdrive.SimDriveClassic(cyc1, veh1)
sim1.sim_drive()
r1=zip(cyc1.cycSecs,sim1.soc*100)

sim2 = simdrive.SimDriveClassic(cyc1, veh2)
sim2.sim_drive()
r2=zip(cyc1.cycSecs,sim2.soc*100)

sim3 = simdrive.SimDriveClassic(cyc1, veh3)
sim3.sim_drive()
r3=zip(cyc1.cycSecs,sim3.soc*100)
```

```python
plt.plot(cyc1.cycSecs,sim1.soc*100,label='Camera only')
plt.plot(cyc1.cycSecs,sim2.soc*100,label='Camera-LiDAR Fusion')
plt.plot(cyc1.cycSecs,sim3.soc*100,label='Camera-Radar Fusion')
plt.ylabel("State of Charge (%)")
plt.xlabel('Time (sec)')
plt.title('SOC (%) Vs Time (sec) for UDDS cycle')
plt.legend()
print(veh1.Scenario_name)
```
