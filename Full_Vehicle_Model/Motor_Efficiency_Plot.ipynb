```python
import numpy as np
import pandas as pd
from scipy.interpolate import LinearNDInterpolator
from matplotlib import pyplot as plt
```

```python
front = pd.read_csv("Audi_e_Tron_Motors_ResultsForMaps.xlsx - APA250-Maps.csv")
front
```

```python
x = front["SPEED [rad/s]"]
y = front["TORQUE [Nm]"]
z = front["EFFY [%]"]
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
ax.view_init(elev=90, azim=-90)

surf = ax.plot_trisurf(x, y, z, cmap="viridis", linewidth=0)
ax.set_xlabel('Speed (rad/s)', fontsize=15, rotation=150)
ax.set_ylabel('Torque (Nm)',fontsize = 15, rotation = 150)
```
