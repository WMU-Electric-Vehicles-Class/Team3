## This python code will be used to determine the electric vehicle's Power and Torque with respect to motor speed. The inputs will be different angular velocities of the motor and the output will be the torque.

> ### Motor Control Signal (%) = Xm
> ### Base Motor Speed (rad/sec) = BMS
> ### Constant Motor Power (W) = P_cons
> ### Max Motor Speed (rad/sec) = 350 


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def piecewise (Wm):
    if 0 <= Wm <= BMS:
        return Xm*Tmstar
    elif Wm >= BMS and Wm <= Wmax:
        return Xm*P_cons/Wm
```

```python
Xm=np.array([1,0.75,0.5,0.25])
P_cons = 64000
BMS = 160 #assumed from lecture's use
Tmstar=P_cons/BMS 
Wmax=350
```

```python
Wm=np.arange(0,Wmax,1)
y=[]
for i in range(len(Wm)):
    y.append(piecewise(Wm[i]))
```

```python
plt.plot(Wm,y)
plt.title("Torque (N-m) at different Motor Control Signal levels")
plt.xlabel('Angular Velocity of Motor (rad/s)')
plt.ylabel('Torque (N-m)')
plt.ylim((0,425))
plt.text(100,405,'100% Xm')
plt.text(100,305,'75% Xm')
plt.text(100,205,'50% Xm')
plt.text(100,105,'25% Xm')
plt.grid()
plt.savefig('MotorPlot1.png',dpi = 1000)
plt.show()
```
