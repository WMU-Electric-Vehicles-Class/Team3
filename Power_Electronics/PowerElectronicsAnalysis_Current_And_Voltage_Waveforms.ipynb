## This code is used for power electronics analysis. The input will be time and the output will be current and voltage at that specific time. 
> ### Frequency (Hz) = freq
> ### Nominal Voltage (V) = V
> ### Discharge Rate (A) = I
> ### Time (sec) = t

```python
import numpy as np
from scipy import signal as sg
import matplotlib.pyplot as plt
```

```python
freq = 50
V=356
I=(10500/240)
t = np.linspace(0, 20, 500)
```

```python
signal1 = V*np.sin(2*np.pi*freq*t)
plt.figure(figsize=(10,4))
plt.plot(t, signal1)
plt.title("Sinusoidal Voltage Signal for Single Phase Power Electronics System")
plt.xlabel('Time (s)')
plt.ylabel('Voltage (V)')
plt.grid()
plt.savefig('PowerElectronicsPlot1.png',dpi=1000)
plt.show()
```

```python
signal2 = I*np.sin(2*np.pi*freq*t)
plt.figure(figsize=(10,4))
plt.plot(t, signal2)
plt.title("Sinusoidal Current Signal for Single Phase Power Electronics System")
plt.xlabel('Time (s)')
plt.ylabel('Current (A)')
plt.grid()
plt.savefig('PowerElectronicsPlot2.png',dpi=1000)
plt.show()
```
