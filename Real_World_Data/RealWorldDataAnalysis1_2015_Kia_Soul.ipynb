## This code is used for real world data analysis. Using Argonne National Lab's Dynamometer Database, the data for a UDDS Hot Start drive-cycle was analyzed. SOC and Voltage over time were plotted for the 2015 Kia Soul Electric.

```python
import matplotlib.pyplot as plt
 
x = []
y = []
z=[]
for line in open('61506048 Test Data.txt', 'r'):
    lines = [i for i in line.split()]
    x.append(float(lines[0]))
    y.append(float(lines[9]))
    z.append (float(lines[8]))
fig,ax = plt.subplots(figsize=(9,5))
ax.plot(x,y,label='SOC',color='xkcd:blue')
ax.grid()
ax2 = ax.twinx()
speed_line = ax2.plot(x,z,label='Speed',color='xkcd:pale red')
ax2.grid(False)
ax.set_xlabel('Time (sec)',fontsize=15)
ax.set_ylabel('SOC (%)',color='xkcd:blue',fontsize =15)
ax.tick_params('y',colors='xkcd:blue')
ax2.set_ylabel('Voltage (V)',color='xkcd:pale red',fontsize=15)
ax2.tick_params('y',colors='xkcd:pale red')
ax.set_title('SOC (%) and Voltage (V) over Time (sec) for a UDDS Hot Start.',fontsize=15)
plt.savefig('RealWorldDataPlot11.png',dpi=1000)
plt.show()
```
