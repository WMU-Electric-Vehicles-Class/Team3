## This code is used for real world data analysis. Using Argonne National Lab's Dynamometer Database, the data for a UDDS Hot Start drive-cycle was analyzed. Current and speed over time were plotted for the 2015 Kia Soul Electric.

```python
import matplotlib.pyplot as plt
 
a = []
b = []
c=[]
for line in open('61506048 Test Data.txt', 'r'):
    lines = [i for i in line.split()]
    a.append(float(lines[0]))
    b.append(float(lines[7]))
    c.append (float(lines[4]))
fig,ax = plt.subplots(figsize=(9,5))
ax.plot(a,b,label='Current',color='xkcd:blue')
ax.grid()
ax2 = ax.twinx()
speed_line = ax2.plot(a,c,label='Speed',color='xkcd:pale red')
ax2.grid(False)
ax.set_xlabel('Time (sec)',fontsize=15)
ax.set_ylabel('Current (A)',color='xkcd:blue',fontsize=15)
ax.tick_params('y',colors='xkcd:blue')
ax2.set_ylabel('Speed (mph)',color='xkcd:pale red',fontsize=15)
ax2.tick_params('y',colors='xkcd:pale red')
ax.set_title('Current (A) and Speed (mph) over Time (sec) for a UDDS Hot Start.',fontsize=15)
plt.savefig('RealWorldDataPlot12.png',dpi=1000)
plt.show()
```
